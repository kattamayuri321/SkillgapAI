```python
from sentence_transformers import SentenceTransformer, util
import pandas as pd

# Load SBERT model
model = SentenceTransformer('all-MiniLM-L6-v2')

# Resume skills
resume_skills = ["python", "sql", "excel", "data analysis", "power bi"]

# Job description skills
job_description_skills = ["python", "sql", "data analytics", "machine learning", "tableau"]

# Generate embeddings
resume_embeddings = model.encode(resume_skills, convert_to_tensor=True)
job_embeddings = model.encode(job_description_skills, convert_to_tensor=True)

# Compute similarity matrix
similarity_matrix = util.cos_sim(resume_embeddings, job_embeddings)

# Convert to DataFrame
similarity_df = pd.DataFrame(
    similarity_matrix.cpu().numpy(),
    index=resume_skills,
    columns=job_description_skills
)

# Find best match and similarity score for each JD skill
best_matches = similarity_df.idxmax(axis=0)
best_scores = similarity_df.max(axis=0)

# Define thresholds
matched_threshold = 0.8
partial_threshold = 0.5

# Classify skills
def classify(score):
    if score >= matched_threshold:
        return "Matched"
    elif score >= partial_threshold:
        return "Partially Matched"
    else:
        return "Missing"

classification = best_scores.apply(classify)

# Combine results
final_df = pd.DataFrame({
    "Job Description Skill": job_description_skills,
    "Best Matching Resume Skill": best_matches,
    "Similarity Score": best_scores.round(3),
    "Classification": classification
})

# Display final skill matching
print(final_df)
```

```python
from sentence_transformers import SentenceTransformer, util
import pandas as pd

# Resume and job description skills
resume_skills = ["Python", "SQL", "Machine Learning", "Deep Learning", "Power BI", "AI"]
job_description_skills = ["Python", "SQL", "Machine Learning", "Deep Learning", "Tableau", "Artificial Intelligence"]

# Normalize abbreviations
abbreviation_map = {"ML":"Machine Learning", "DL":"Deep Learning", "AI":"Artificial Intelligence"}
def normalize_skill(skill):
    skill = skill.strip()
    skill = skill.upper()
    return abbreviation_map.get(skill, skill)

normalized_resume_skills = [normalize_skill(s) for s in resume_skills]
normalized_job_skills = [normalize_skill(s) for s in job_description_skills]

# Load two different SBERT models
model_fast = SentenceTransformer('all-MiniLM-L6-v2')     # Fast, lightweight
model_high_acc = SentenceTransformer('all-mpnet-base-v2') # Higher accuracy

# Generate embeddings
emb_fast_resume = model_fast.encode(normalized_resume_skills, convert_to_tensor=True)
emb_fast_job = model_fast.encode(normalized_job_skills, convert_to_tensor=True)

emb_acc_resume = model_high_acc.encode(normalized_resume_skills, convert_to_tensor=True)
emb_acc_job = model_high_acc.encode(normalized_job_skills, convert_to_tensor=True)

# Compute similarity matrices
sim_fast = util.cos_sim(emb_fast_resume, emb_fast_job)
sim_acc = util.cos_sim(emb_acc_resume, emb_acc_job)

# Convert to DataFrames
df_fast = pd.DataFrame(sim_fast.cpu().numpy(), index=normalized_resume_skills, columns=normalized_job_skills).round(3)
df_acc = pd.DataFrame(sim_acc.cpu().numpy(), index=normalized_resume_skills, columns=normalized_job_skills).round(3)

# Display both similarity matrices
print("=== Similarity Matrix (Fast Model: all-MiniLM-L6-v2) ===")
display(df_fast)
print("\n=== Similarity Matrix (High Accuracy Model: all-mpnet-base-v2) ===")
display(df_acc)
```

```python
from sentence_transformers import SentenceTransformer, util
import pandas as pd

# Load SBERT model
model = SentenceTransformer('all-MiniLM-L6-v2')

# Example resume and job skills
resume_skills = ["Python", "SQL", "Machine Learning", "Deep Learning", "Power BI", "AI"]
job_description_skills = ["Python", "SQL", "Machine Learning", "Deep Learning", "Tableau", "AI"]

# Abbreviation normalization
abbreviation_map = {"ML":"Machine Learning", "DL":"Deep Learning", "AI":"Artificial Intelligence"}
def normalize_skill(skill):
    skill = skill.strip()
    skill = skill.upper()
    return abbreviation_map.get(skill, skill)

normalized_resume_skills = [normalize_skill(s) for s in resume_skills]
normalized_job_skills = [normalize_skill(s) for s in job_description_skills]

# Build a set of unique skills to avoid repeated embeddings
unique_skills = list(set(normalized_resume_skills + normalized_job_skills))

# Cache embeddings in a dictionary
embedding_cache = {}
for skill in unique_skills:
    embedding_cache[skill] = model.encode(skill, convert_to_tensor=True)

# Function to get embeddings from cache
def get_embedding(skill):
    return embedding_cache[skill]

# Generate embeddings for resume and job description using cache
resume_embeddings = [get_embedding(skill) for skill in normalized_resume_skills]
job_embeddings = [get_embedding(skill) for skill in normalized_job_skills]

# Convert lists to tensors for similarity computation
import torch
resume_embeddings_tensor = torch.stack(resume_embeddings)
job_embeddings_tensor = torch.stack(job_embeddings)

# Compute cosine similarity
similarity_matrix = util.cos_sim(resume_embeddings_tensor, job_embeddings_tensor)

# Convert to DataFrame
similarity_df = pd.DataFrame(
    similarity_matrix.cpu().numpy(),
    index=normalized_resume_skills,
    columns=normalized_job_skills
).round(3)

# Display similarity matrix
print("Similarity Matrix with Cached Embeddings:")
display(similarity_df)
```

```python
# Full Skill Matching Pipeline: Raw Resume vs Job Description
from sentence_transformers import SentenceTransformer, util
import pandas as pd
import json
import seaborn as sns
import matplotlib.pyplot as plt
import re
import torch

# ------------------------------
# 1. Example raw texts
# ------------------------------
raw_resume_text = """
Experienced in Python, SQL, ML, DL, Power BI, AI.
Worked on data pipelines and analytics.
"""

raw_jd_text = """
Looking for candidates skilled in Python, SQL, Machine Learning, Deep Learning,
Tableau, and Artificial Intelligence.
"""

# ------------------------------
# 2. Extract skills from text
# Simple approach: split by commas and spaces
# ------------------------------
def extract_skills(text):
    # remove punctuation except + and #
    text_clean = re.sub(r'[^\w\s+#]', '', text)
    # split by spaces and commas
    words = re.split(r'[\s,]+', text_clean)
    # remove empty strings and duplicates
    return list(set([w for w in words if w]))

resume_skills_raw = extract_skills(raw_resume_text)
jd_skills_raw = extract_skills(raw_jd_text)

# ------------------------------
# 3. Normalize skills (case + abbreviations)
# ------------------------------
abbreviation_map = {
    "ML": "Machine Learning",
    "DL": "Deep Learning",
    "AI": "Artificial Intelligence"
}

def normalize_skill(skill):
    skill = skill.strip()
    skill_upper = skill.upper()
    return abbreviation_map.get(skill_upper, skill.title())

resume_skills = [normalize_skill(s) for s in resume_skills_raw]
jd_skills = [normalize_skill(s) for s in jd_skills_raw]

# Remove duplicates
resume_skills = list(set(resume_skills))
jd_skills = list(set(jd_skills))

print("Normalized Resume Skills:", resume_skills)
print("Normalized JD Skills:", jd_skills)

# ------------------------------
# 4. Load SBERT model
# ------------------------------
model = SentenceTransformer('all-MiniLM-L6-v2')

# ------------------------------
# 5. Cache embeddings for unique skills
# ------------------------------
unique_skills = list(set(resume_skills + jd_skills))
embedding_cache = {skill: model.encode(skill, convert_to_tensor=True) for skill in unique_skills}

def get_embedding(skill):
    return embedding_cache[skill]

resume_embeddings = torch.stack([get_embedding(s) for s in resume_skills])
jd_embeddings = torch.stack([get_embedding(s) for s in jd_skills])

# ------------------------------
# 6. Compute similarity matrix
# ------------------------------
similarity_matrix = util.cos_sim(resume_embeddings, jd_embeddings)
similarity_df = pd.DataFrame(
    similarity_matrix.cpu().numpy(),
    index=resume_skills,
    columns=jd_skills
).round(3)

print("\n=== Similarity Matrix ===")
display(similarity_df)

# ------------------------------
# 7. Classify matches
# ------------------------------
matched_threshold = 0.8
partial_threshold = 0.5

best_matches = similarity_df.idxmax(axis=0)
best_scores = similarity_df.max(axis=0)

def classify(score):
    if score >= matched_threshold:
        return "Matched"
    elif score >= partial_threshold:
        return "Partially Matched"
    else:
        return "Missing"

classification = best_scores.apply(classify)

skill_gap_report = pd.DataFrame({
    "Job Description Skill": jd_skills,
    "Best Matching Resume Skill": best_matches,
    "Similarity Score": best_scores.round(3),
    "Classification": classification
})

print("\n=== Skill Gap Report ===")
display(skill_gap_report)

# ------------------------------
# 8. Save report as JSON
# ------------------------------
skill_gap_json = skill_gap_report.to_dict(orient="records")
with open("skill_gap_report.json", "w") as f:
    json.dump(skill_gap_json, f, indent=4)
print("\nSkill gap report saved as 'skill_gap_report.json'")

# ------------------------------
# 9. Optional: Visualize similarity as heatmap
# ------------------------------
plt.figure(figsize=(10, 6))
sns.heatmap(similarity_df, annot=True, fmt=".2f", cmap="YlGnBu", cbar=True)
plt.title("Resume vs Job Description Skills Similarity Heatmap", fontsize=16)
plt.xlabel("Job Description Skills", fontsize=12)
plt.ylabel("Resume Skills", fontsize=12)
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
from sentence_transformers import SentenceTransformer, util
import pandas as pd
import torch

# ------------------------------
# Example normalized skills
# ------------------------------
resume_skills = ["Python", "SQL", "Machine Learning", "Communication", "Leadership", "Teamwork"]
jd_skills = ["Python", "SQL", "Machine Learning", "Communication", "Team Management"]

# Define which skills are technical vs soft
technical_skills = ["Python", "SQL", "Machine Learning"]
soft_skills = ["Communication", "Leadership", "Teamwork", "Team Management"]

# Load SBERT model
model = SentenceTransformer('all-MiniLM-L6-v2')

# ------------------------------
# Cache embeddings
# ------------------------------
unique_skills = list(set(resume_skills + jd_skills))
embedding_cache = {skill: model.encode(skill, convert_to_tensor=True) for skill in unique_skills}

def get_embedding(skill):
    return embedding_cache[skill]

resume_embeddings = torch.stack([get_embedding(skill) for skill in resume_skills])
jd_embeddings = torch.stack([get_embedding(skill) for skill in jd_skills])

# ------------------------------
# Compute similarity
# ------------------------------
similarity_matrix = util.cos_sim(resume_embeddings, jd_embeddings)
similarity_df = pd.DataFrame(
    similarity_matrix.cpu().numpy(),
    index=resume_skills,
    columns=jd_skills
).round(3)

print("=== Similarity Matrix ===")
display(similarity_df)

# ------------------------------
# Apply thresholds
# ------------------------------
# Example thresholds:
tech_matched_threshold = 0.8
tech_partial_threshold = 0.6
soft_matched_threshold = 0.7
soft_partial_threshold = 0.5

top_matches = similarity_df.idxmax(axis=0)
top_scores = similarity_df.max(axis=0)

def classify(skill, score):
    if skill in technical_skills:
        if score >= tech_matched_threshold:
            return "Matched"
        elif score >= tech_partial_threshold:
            return "Partially Matched"
        else:
            return "Missing"
    elif skill in soft_skills:
        if score >= soft_matched_threshold:
            return "Matched"
        elif score >= soft_partial_threshold:
            return "Partially Matched"
        else:
            return "Missing"
    else:
        # Default to technical thresholds if unknown
        if score >= tech_matched_threshold:
            return "Matched"
        elif score >= tech_partial_threshold:
            return "Partially Matched"
        else:
            return "Missing"

classification = pd.Series({jd: classify(jd, score) for jd, score in top_scores.items()})

# ------------------------------
# Build final skill gap report
# ------------------------------
skill_gap_report = pd.DataFrame({
    "Job Description Skill": jd_skills,
    "Best Matching Resume Skill": top_matches,
    "Similarity Score": top_scores.round(3),
    "Classification": classification
})

print("\n=== Skill Gap Report with Different Thresholds ===")
display(skill_gap_report)
```

```python
from sentence_transformers import SentenceTransformer, util
import pandas as pd
import torch
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages
import json
class EmbeddingGenerator:
    def __init__(self, model_name='all-MiniLM-L6-v2'):
        self.model = SentenceTransformer(model_name)
        self.cache = {}

    def get_embedding(self, skill):
        if skill not in self.cache:
            self.cache[skill] = self.model.encode(skill, convert_to_tensor=True)
        return self.cache[skill]

    def encode_list(self, skills):
        return torch.stack([self.get_embedding(s) for s in skills])
class SimilarityCalculator:
    def __init__(self, resume_skills, jd_skills, embedding_generator):
        self.resume_skills = resume_skills
        self.jd_skills = jd_skills
        self.emb_gen = embedding_generator
        self.similarity_df = None

    def compute_similarity(self):
        resume_embeddings = self.emb_gen.encode_list(self.resume_skills)
        jd_embeddings = self.emb_gen.encode_list(self.jd_skills)
        similarity_matrix = util.cos_sim(resume_embeddings, jd_embeddings)
        self.similarity_df = pd.DataFrame(
            similarity_matrix.cpu().numpy(),
            index=self.resume_skills,
            columns=self.jd_skills
        ).round(3)
        return self.similarity_df

    def top_n_matches(self, n=3):
        top_matches = {}
        for jd in self.jd_skills:
            sorted_scores = self.similarity_df[jd].sort_values(ascending=False)
            top_matches[jd] = sorted_scores.head(n).to_dict()
        return top_matches
class SkillGapReporter:
    def __init__(self, similarity_df, matched_threshold=0.8, partial_threshold=0.5):
        self.similarity_df = similarity_df
        self.matched_threshold = matched_threshold
        self.partial_threshold = partial_threshold
        self.report = None

    def classify_score(self, score):
        if score >= self.matched_threshold:
            return "Matched"
        elif score >= self.partial_threshold:
            return "Partially Matched"
        else:
            return "Missing"

    def generate_report(self):
        best_matches = self.similarity_df.idxmax(axis=0)
        best_scores = self.similarity_df.max(axis=0)
        classification = best_scores.apply(self.classify_score)
        self.report = pd.DataFrame({
            "Job Description Skill": self.similarity_df.columns,
            "Best Matching Resume Skill": best_matches,
            "Similarity Score": best_scores.round(3),
            "Classification": classification
        })
        return self.report

    def save_json(self, filename="skill_gap_report.json"):
        if self.report is not None:
            with open(filename, "w") as f:
                json.dump(self.report.to_dict(orient="records"), f, indent=4)
            print(f"Report saved as {filename}")

    def plot_heatmap(self):
        plt.figure(figsize=(10, 6))
        sns.heatmap(self.similarity_df, annot=True, fmt=".2f", cmap="YlGnBu", cbar=True)
        plt.title("Resume vs Job Description Skills Similarity Heatmap", fontsize=14, fontweight='bold')
        plt.xlabel("Job Description Skills")
        plt.ylabel("Resume Skills")
        plt.xticks(rotation=45)
        plt.yticks(rotation=0)
        plt.tight_layout()
        plt.show()

# Skills
resume_skills = ["Python", "SQL", "Machine Learning", "Deep Learning", "Power BI", "AI"]
jd_skills = ["Python", "SQL", "Machine Learning", "Deep Learning", "Tableau", "AI"]

# Embedding Generation
emb_gen = EmbeddingGenerator(model_name='all-MiniLM-L6-v2')

# Similarity Computation
sim_calc = SimilarityCalculator(resume_skills, jd_skills, emb_gen)
similarity_df = sim_calc.compute_similarity()
print("=== Similarity Matrix ===")
display(similarity_df)

# Top 3 Matches per JD skill
top_matches = sim_calc.top_n_matches(n=3)
print("\n=== Top 3 Resume Skills for Each JD Skill ===")
for jd, matches in top_matches.items():
    print(f"{jd}: {matches}")

# Skill Gap Reporting
reporter = SkillGapReporter(similarity_df, matched_threshold=0.8, partial_threshold=0.5)
skill_gap_report = reporter.generate_report()
print("\n=== Skill Gap Report ===")
display(skill_gap_report)

# Save report as JSON
reporter.save_json()

# Plot heatmap
reporter.plot_heatmap()
```
