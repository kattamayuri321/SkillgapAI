```python
resume_text = "Experienced in Python, SQL, and Java with basic knowledge of HTML."

skill_list = ["python", "java", "c", "sql", "html"]
found_skills = []

for skill in skill_list:
    if skill in resume_text.lower():
        found_skills.append(skill.capitalize())

print("Extracted Technical Skills:")
for skill in found_skills:
    print(skill)
```

```python
# Sample resume text
resume_text = "Python SQL Teamwork Python Java SQL Python Communication"

# Predefined skill list
skills = ['Python', 'Java', 'SQL', 'Teamwork', 'Communication', 'Leadership']

# Convert resume text to lowercase and split into words
words = resume_text.lower().split()

# Count occurrences of each skill
skill_count = {}
for skill in skills:
    skill_count[skill] = words.count(skill.lower())

print(skill_count)
```

```python
import csv

# Sample resume text
resume_text = "Experienced in python, SQL, and teamwork"

# Convert resume text to lowercase for case-insensitive matching
resume_words = resume_text.lower().split()

# Load skills from CSV file
skills = []
with open('skills.csv', 'r') as file:
    reader = csv.reader(file)
    for row in reader:
        for skill in row:
            skills.append(skill.strip().lower())  # Convert to lowercase

# Find matching skills
matching_skills = [skill for skill in skills if skill in resume_words]

print("Matching skills:", matching_skills)
```

```python
import csv

# Example extracted skills
extracted_skills = ['Python', 'SQL', 'Teamwork', 'Java', 'Communication']

# Name of the new CSV file
output_file = 'extracted_skills.csv'

# Write skills to CSV
with open(output_file, 'w', newline='') as file:
    writer = csv.writer(file)
    # Write each skill in a separate row
    for skill in extracted_skills:
        writer.writerow([skill])

print(f"Extracted skills have been saved to {output_file}")
```
