```python
# Install necessary libraries
!pip install PyPDF2 python-docx

import os
from PyPDF2 import PdfReader
from docx import Document

def load_file(path):
    """
    Load text from PDF, DOCX, or TXT files.
    Returns the extracted text as a string.
    """
    if not os.path.exists(path):
        raise FileNotFoundError(f"File '{path}' not found.")

    _, ext = os.path.splitext(path)
    ext = ext.lower()
    text = ""

    try:
        if ext == ".txt":
            with open(path, "r", encoding="utf-8") as f:
                text = f.read()
        elif ext == ".pdf":
            with open(path, "rb") as f:
                reader = PdfReader(f)
                for page in reader.pages:
                    page_text = page.extract_text()
                    if page_text:  # Some pages may return None
                        text += page_text + "\n"
        elif ext == ".docx":
            doc = Document(path)
            text = "\n".join([para.text for para in doc.paragraphs if para.text.strip() != ""])
        else:
            raise ValueError(f"Unsupported file type: '{ext}'")
    except Exception as e:
        raise RuntimeError(f"Error reading file '{path}': {e}")

    return text

# Example usage
try:
    resume_text = load_file("resume.pdf")  # replace with your file path
    job_desc_text = load_file("job_description.txt")  # replace with your file path

    print("Resume text loaded successfully. Preview:\n")
    print(resume_text[:500], "...\n")

    print("Job description text loaded successfully. Preview:\n")
    print(job_desc_text[:500], "...")
except Exception as e:
    print(e)
```

```python
# Enhanced unified loader with error handling
import os
from PyPDF2 import PdfReader
from docx import Document

def load_file(path):
    """
    Load text from PDF, DOCX, or TXT files.
    Raises appropriate errors for unsupported types, empty files, or invalid paths.
    """
    if not os.path.exists(path):
        raise FileNotFoundError(f"File '{path}' does not exist.")

    _, ext = os.path.splitext(path)
    ext = ext.lower()
    text = ""

    try:
        if ext == ".txt":
            with open(path, "r", encoding="utf-8") as f:
                text = f.read()
        elif ext == ".pdf":
            with open(path, "rb") as f:
                reader = PdfReader(f)
                for page in reader.pages:
                    page_text = page.extract_text()
                    if page_text:
                        text += page_text + "\n"
        elif ext == ".docx":
            doc = Document(path)
            text = "\n".join([para.text for para in doc.paragraphs if para.text.strip() != ""])
        else:
            raise ValueError(f"Unsupported file type '{ext}'. Supported types: PDF, DOCX, TXT.")

        if not text.strip():
            raise ValueError(f"The file '{path}' is empty or contains no readable text.")

    except Exception as e:
        raise RuntimeError(f"Error reading file '{path}': {e}")

    return text


# Simulate file upload with error handling
resume_path = input("Enter the path to your resume file (PDF, DOCX, or TXT): ")
job_desc_path = input("Enter the path to your job description file (PDF, DOCX, or TXT): ")

try:
    resume_text = load_file(resume_path)
    job_desc_text = load_file(job_desc_path)

    print("\n✅ Files loaded successfully!")

    # Display previews
    print("\n=== Resume Preview (first 500 chars) ===\n")
    print(resume_text[:500], "...\n")

    print("=== Job Description Preview (first 500 chars) ===\n")
    print(job_desc_text[:500], "...\n")

except FileNotFoundError as e:
    print(f"❌ File not found: {e}")
except ValueError as e:
    print(f"❌ Value error: {e}")
except RuntimeError as e:
    print(f"❌ Runtime error: {e}")
except Exception as e:
    print(f"❌ An unexpected error occurred: {e}")
```

```python
# Install necessary libraries
!pip install PyPDF2 python-docx

import os
from PyPDF2 import PdfReader
from docx import Document

def universal_parser(file_path, preview_chars=500):
    """
    Detect file type, extract text, clean it, and display preview.

    Returns a dictionary with:
        - 'raw': original extracted text
        - 'cleaned': cleaned text
        - 'preview': first N characters of cleaned text
        - 'file_path': original file path
    """
    if not os.path.exists(file_path):
        raise FileNotFoundError(f"File '{file_path}' not found.")

    _, ext = os.path.splitext(file_path)
    ext = ext.lower()
    text = ""

    try:
        # Extract text based on file type
        if ext == ".txt":
            with open(file_path, "r", encoding="utf-8") as f:
                text = f.read()
        elif ext == ".pdf":
            with open(file_path, "rb") as f:
                reader = PdfReader(f)
                for page in reader.pages:
                    page_text = page.extract_text()
                    if page_text:
                        text += page_text + "\n"
        elif ext == ".docx":
            doc = Document(file_path)
            text = "\n".join([para.text for para in doc.paragraphs if para.text.strip() != ""])
        else:
            raise ValueError(f"Unsupported file type '{ext}'. Supported types: PDF, DOCX, TXT.")

        if not text.strip():
            raise ValueError(f"The file '{file_path}' is empty or contains no readable text.")

        # Clean text
        cleaned_text = ' '.join(text.lower().split())

        # Create preview
        preview = cleaned_text[:preview_chars]

        # Display preview
        print(f"\n=== Preview of '{file_path}' (first {preview_chars} chars) ===\n")
        print(preview, "...\n")

        return {
            "file_path": file_path,
            "raw": text,
            "cleaned": cleaned_text,
            "preview": preview
        }

    except Exception as e:
        raise RuntimeError(f"Error processing file '{file_path}': {e}")


# Example usage:
try:
    resume_data = universal_parser("resume.pdf")  # replace with your file path
    job_desc_data = universal_parser("job_description.txt")  # replace with your file path

except Exception as e:
    print(f"An error occurred: {e}")
```

```python
# List of multiple resume file paths (PDF, DOCX, TXT)
resume_files = [
    "resume.pdf",
    "sample_resume.docx",
    "resume.txt"
]

# Single job description file path
job_desc_file = "job_description.txt"

# List to store parsed resumes
parsed_resumes = []

try:
    # Parse each resume using the universal parser
    for path in resume_files:
        try:
            resume_data = universal_parser(path)
            parsed_resumes.append(resume_data)
        except Exception as e:
            print(f"❌ Error parsing '{path}': {e}")

    # Parse the job description
    job_desc_data = universal_parser(job_desc_file)

    print("\n✅ All files parsed successfully!\n")

    # Display previews for each resume
    for i, resume in enumerate(parsed_resumes):
        print(f"=== Resume {i+1} Preview ({resume['file_path']}) ===")
        print(resume['preview'], "...\n")

    # Display preview for the job description
    print(f"=== Job Description Preview ({job_desc_data['file_path']}) ===")
    print(job_desc_data['preview'], "...\n")

except Exception as e:
    print(f"An error occurred during parsing: {e}")
```
